### ðŸŽ¯ Exemplo 2: FunÃ§Ã£o com GrÃ¡fico


````python


def grafico_funcao_quadratica(a, b, c):
    '''
    Cria grÃ¡fico de uma funÃ§Ã£o quadrÃ¡tica y = axÂ² + bx + c
    '''
    # Criar pontos para o grÃ¡fico
    x = np.linspace(-10, 10, 200)
    y = a * x**2 + b * x + c

    # Criar o grÃ¡fico
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'r-', linewidth=3, label=f'y = {a}xÂ² + {b}x + {c}')

    # Destacar o vÃ©rtice
    x_vertice = -b / (2*a)
    y_vertice = a * x_vertice**2 + b * x_vertice + c
    plt.plot(x_vertice, y_vertice, 'go', markersize=10, label=f'VÃ©rtice: ({x_vertice:.1f}, {y_vertice:.1f})')

    # Configurar o grÃ¡fico
    plt.title(f'FunÃ§Ã£o QuadrÃ¡tica: y = {a}xÂ² + {b}x + {c}', fontsize=14, fontweight='bold')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.axhline(y=0, color='k', linewidth=0.8)
    plt.axvline(x=0, color='k', linewidth=0.8)
    plt.show()

    # Mostrar anÃ¡lise
    print(f"ðŸ“Š AnÃ¡lise da funÃ§Ã£o y = {a}xÂ² + {b}x + {c}:")
    print(f"   â€¢ Coeficiente a: {a} ({'positivo' if a > 0 else 'negativo'} - concavidade {'para cima' if a > 0 else 'para baixo'})")
    print(f"   â€¢ VÃ©rtice: ({x_vertice:.2f}, {y_vertice:.2f})")

    if a > 0:
        print(f"   â€¢ Valor mÃ­nimo: y = {y_vertice:.2f}")
    else:
        print(f"   â€¢ Valor mÃ¡ximo: y = {y_vertice:.2f}")

# Testando nossa funÃ§Ã£o
grafico_funcao_quadratica(1, -2, -3)
````

### ðŸŽ¯ Exemplo 1: Slider Simples

````python

def funcao_com_slider(a=1):
    '''
    FunÃ§Ã£o que serÃ¡ controlada por um slider
    '''
    x = np.linspace(-5, 5, 100)
    y = a * x

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, 'b-', linewidth=3, label=f'y = {a}x')
    plt.title(f'FunÃ§Ã£o Linear: y = {a}x', fontsize=14)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    plt.ylim(-10, 10)
    plt.legend()
    plt.show()

# Criar o elemento interativo
interact(funcao_com_slider,
         a=FloatSlider(min=-3, max=3, step=0.1, value=1, description='Coeficiente a:'))
````


### ðŸŽ¯ Exemplo 2: MÃºltiplos Controles


````python


def explorar_parabola(a=1, b=0, c=0, mostrar_vertice=True, cor='azul'):
    '''
    Explora parÃ¡bolas com mÃºltiplos controles
    '''
    x = np.linspace(-10, 10, 200)
    y = a * x**2 + b * x + c

    # Definir cor
    cores = {'azul': 'blue', 'vermelho': 'red', 'verde': 'green', 'roxo': 'purple'}
    cor_grafico = cores[cor]

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, color=cor_grafico, linewidth=3, label=f'y = {a}xÂ² + {b}x + {c}')

    if mostrar_vertice:
        x_v = -b / (2*a)
        y_v = a * x_v**2 + b * x_v + c
        plt.plot(x_v, y_v, 'ro', markersize=10, label=f'VÃ©rtice: ({x_v:.1f}, {y_v:.1f})')

    plt.title(f'y = {a}xÂ² + {b}x + {c}', fontsize=14, fontweight='bold')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.axhline(y=0, color='k', linewidth=0.8)
    plt.axvline(x=0, color='k', linewidth=0.8)
    plt.ylim(-20, 20)
    plt.show()

# Interface interativa completa
interact(explorar_parabola,
         a=FloatSlider(min=-2, max=2, step=0.1, value=1, description='a (xÂ²):'),
         b=FloatSlider(min=-5, max=5, step=0.5, value=0, description='b (x):'),
         c=FloatSlider(min=-10, max=10, step=1, value=0, description='c (constante):'),
         mostrar_vertice=Checkbox(value=True, description='Mostrar vÃ©rtice'),
         cor=Dropdown(options=['azul', 'vermelho', 'verde', 'roxo'], value='azul', description='Cor:'))
````

### ðŸ“š Exemplo: Mini-Aula sobre FunÃ§Ã£o Exponencial

````python

# ===== INTRODUÃ‡ÃƒO =====
print("ðŸŽ¯ MINI-AULA: FUNÃ‡ÃƒO EXPONENCIAL")
print("=" * 50)
print("Objetivo: Compreender o comportamento da funÃ§Ã£o f(x) = a^x")
print("CompetÃªncias BNCC: EM13MAT302, EM13MAT402")
print()

# ===== CONCEITOS TEÃ“RICOS =====
print("ðŸ“– CONCEITOS FUNDAMENTAIS")
print("-" * 30)
print("â€¢ FunÃ§Ã£o exponencial: f(x) = a^x, onde a > 0 e a â‰  1")
print("â€¢ Se a > 1: funÃ§Ã£o crescente")
print("â€¢ Se 0 < a < 1: funÃ§Ã£o decrescente")
print("â€¢ DomÃ­nio: todos os nÃºmeros reais")
print("â€¢ ContradomÃ­nio: nÃºmeros reais positivos")
print()

# ===== EXEMPLO INTERATIVO =====
def explorar_exponencial(base=2, mostrar_grafico_log=False, incluir_comparacao=True):
    '''
    Explora funÃ§Ãµes exponenciais interativamente
    '''
    x = np.linspace(-3, 3, 100)
    y = base ** x

    plt.figure(figsize=(12, 6))

    if incluir_comparacao:
        # Mostrar mÃºltiplas bases para comparaÃ§Ã£o
        bases = [0.5, 2, 3]
        cores = ['red', 'blue', 'green']
        for i, b in enumerate(bases):
            y_comp = b ** x
            estilo = '-' if b == base else '--'
            alpha = 1.0 if b == base else 0.5
            plt.plot(x, y_comp, color=cores[i], linestyle=estilo,
                    linewidth=3, alpha=alpha, label=f'y = {b}^x')
    else:
        plt.plot(x, y, 'b-', linewidth=3, label=f'y = {base}^x')

    # Destacar pontos importantes
    plt.plot(0, 1, 'ro', markersize=10, label='(0, 1) - sempre passa aqui!')
    plt.plot(1, base, 'go', markersize=8, label=f'(1, {base})')

    plt.title(f'FunÃ§Ã£o Exponencial: y = {base}^x', fontsize=14, fontweight='bold')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.axhline(y=0, color='k', linewidth=0.8)
    plt.axvline(x=0, color='k', linewidth=0.8)
    plt.ylim(0, 10)

    if mostrar_grafico_log:
        # Adicionar funÃ§Ã£o logarÃ­tmica (inversa)
        y_log = np.log(x[x > 0]) / np.log(base)
        x_log = x[x > 0]
        plt.plot(x_log, y_log, 'm--', linewidth=2, label=f'y = log_{base}(x)')
        plt.legend()

    plt.show()

    # AnÃ¡lise textual
    print(f"ðŸ“Š ANÃLISE DA FUNÃ‡ÃƒO y = {base}^x:")
    if base > 1:
        print(f"   âœ… FunÃ§Ã£o CRESCENTE (base {base} > 1)")
        print("   â€¢ Quando x aumenta, y aumenta rapidamente")
        print("   â€¢ Quando x â†’ -âˆž, y â†’ 0")
        print("   â€¢ Quando x â†’ +âˆž, y â†’ +âˆž")
    elif base < 1:
        print(f"   âœ… FunÃ§Ã£o DECRESCENTE (base {base} < 1)")
        print("   â€¢ Quando x aumenta, y diminui")
        print("   â€¢ Quando x â†’ -âˆž, y â†’ +âˆž")
        print("   â€¢ Quando x â†’ +âˆž, y â†’ 0")

    print(f"   â€¢ Passa sempre pelo ponto (0, 1)")
    print(f"   â€¢ Passa pelo ponto (1, {base})")
    print(f"   â€¢ NÃ£o toca o eixo x (assÃ­ntota horizontal y = 0)")

print("ðŸŽ® EXPLORADOR INTERATIVO:")
interact(explorar_exponencial,
         base=FloatSlider(min=0.1, max=5, step=0.1, value=2, description='Base (a):'),
         mostrar_grafico_log=Checkbox(value=False, description='Mostrar log (funÃ§Ã£o inversa)'),
         incluir_comparacao=Checkbox(value=True, description='Comparar com outras bases'))
````

````python


def exemplo_multiplos_graficos():
    '''
    Como criar mÃºltiplos grÃ¡ficos lado a lado
    '''
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    x = np.linspace(-5, 5, 100)

    # GrÃ¡fico 1
    ax1.plot(x, x**2, 'b-', linewidth=2)
    ax1.set_title('FunÃ§Ã£o QuadrÃ¡tica')
    ax1.set_xlabel('x')
    ax1.set_ylabel('y = xÂ²')
    ax1.grid(True, alpha=0.3)

    # GrÃ¡fico 2
    ax2.plot(x, x**3, 'r-', linewidth=2)
    ax2.set_title('FunÃ§Ã£o CÃºbica')
    ax2.set_xlabel('x')
    ax2.set_ylabel('y = xÂ³')
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

exemplo_multiplos_graficos()
````

#### 2. **AnotaÃ§Ãµes nos GrÃ¡ficos**


````python


def exemplo_anotacoes():
    '''
    Como adicionar anotaÃ§Ãµes e destaques
    '''
    x = np.linspace(-3, 3, 100)
    y = x**2 - 2*x - 3

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'b-', linewidth=3, label='y = xÂ² - 2x - 3')

    # Encontrar e destacar as raÃ­zes
    raizes_x = [-1, 3]  # RaÃ­zes calculadas
    raizes_y = [0, 0]
    plt.plot(raizes_x, raizes_y, 'ro', markersize=10, label='RaÃ­zes')

    # Destacar o vÃ©rtice
    x_vertice = 1
    y_vertice = -4
    plt.plot(x_vertice, y_vertice, 'go', markersize=10, label='VÃ©rtice')

    # Adicionar anotaÃ§Ãµes
    plt.annotate('Raiz: (-1, 0)', xy=(-1, 0), xytext=(-2, 2),
                arrowprops=dict(arrowstyle='->', color='red'),
                fontsize=12, color='red')

    plt.annotate('VÃ©rtice: (1, -4)', xy=(1, -4), xytext=(2, -2),
                arrowprops=dict(arrowstyle='->', color='green'),
                fontsize=12, color='green')

    plt.title('ParÃ¡bola com Pontos Importantes Destacados', fontsize=14, fontweight='bold')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.axhline(y=0, color='k', linewidth=0.8)
    plt.axvline(x=0, color='k', linewidth=0.8)
    plt.show()

exemplo_anotacoes()
````

## ðŸ“š Parte 7: Exemplos PrÃ¡ticos para Diferentes NÃ­veis

### ðŸŽ¯ Exemplo para 9Âº Ano: FunÃ§Ã£o Linear Simplificada

Vamos criar um material adequado para alunos que estÃ£o vendo funÃ§Ã£o linear pela primeira vez:


````python
def funcao_linear_9ano(coeficiente_angular=1, termo_independente=0,
                       mostrar_pontos=True, mostrar_tabela=False):
    '''
    FunÃ§Ã£o linear adaptada para 9Âº ano - linguagem mais simples
    '''
    # Criar pontos para o grÃ¡fico
    x = np.linspace(-5, 5, 100)
    y = coeficiente_angular * x + termo_independente

    plt.figure(figsize=(10, 7))
    plt.plot(x, y, 'b-', linewidth=4, label=f'y = {coeficiente_angular}x + {termo_independente}')

    if mostrar_pontos:
        # Mostrar alguns pontos especÃ­ficos
        x_pontos = np.array([-3, -1, 0, 1, 3])
        y_pontos = coeficiente_angular * x_pontos + termo_independente
        plt.plot(x_pontos, y_pontos, 'ro', markersize=8)

        # Anotar pontos importantes
        plt.plot(0, termo_independente, 'go', markersize=12,
                label=f'Corta o eixo y em: (0, {termo_independente})')

    plt.title(f'FunÃ§Ã£o do 1Âº grau: y = {coeficiente_angular}x + {termo_independente}',
              fontsize=16, fontweight='bold')
    plt.xlabel('x (entrada)', fontsize=14)
    plt.ylabel('y (saÃ­da)', fontsize=14)
    plt.grid(True, alpha=0.4, linewidth=1)
    plt.legend(fontsize=12)
    plt.axhline(y=0, color='k', linewidth=1)
    plt.axvline(x=0, color='k', linewidth=1)
    plt.ylim(-10, 10)
    plt.show()

    # ExplicaÃ§Ã£o simples
    print("ðŸ” O QUE OBSERVAMOS:")
    print("=" * 40)

    if coeficiente_angular > 0:
        print("ðŸ“ˆ A reta estÃ¡ SUBINDO (crescente)")
        print(f"   â†’ Para cada 1 que x aumenta, y aumenta {coeficiente_angular}")
    elif coeficiente_angular < 0:
        print("ðŸ“‰ A reta estÃ¡ DESCENDO (decrescente)")
        print(f"   â†’ Para cada 1 que x aumenta, y diminui {abs(coeficiente_angular)}")
    else:
        print("âž¡ï¸ A reta estÃ¡ na HORIZONTAL (constante)")
        print("   â†’ y sempre vale a mesma coisa, nÃ£o importa o valor de x")

    print(f"ðŸ“ A reta corta o eixo y no ponto (0, {termo_independente})")

    if mostrar_tabela:
        print("\nðŸ“Š TABELA DE VALORES:")
        print("-" * 20)
        print("x  |  y")
        print("-" * 20)
        for x_val in [-2, -1, 0, 1, 2]:
            y_val = coeficiente_angular * x_val + termo_independente
            print(f"{x_val:2} | {y_val:3}")

print("ðŸŽ® EXPLORE A FUNÃ‡ÃƒO DO 1Âº GRAU:")
print("(Adequado para 9Âº ano)")

interact(funcao_linear_9ano,
         coeficiente_angular=FloatSlider(min=-3, max=3, step=0.5, value=1,
                                       description='InclinaÃ§Ã£o da reta:'),
         termo_independente=FloatSlider(min=-5, max=5, step=1, value=0,
                                      description='Onde corta o eixo y:'),
         mostrar_pontos=Checkbox(value=True, description='Mostrar pontos importantes'),
         mostrar_tabela=Checkbox(value=False, description='Mostrar tabela de valores'))
````
