### üéØ Exemplo 2: Fun√ß√£o com Gr√°fico


````python


def grafico_funcao_quadratica(a, b, c):
    '''
    Cria gr√°fico de uma fun√ß√£o quadr√°tica y = ax¬≤ + bx + c
    '''
    # Criar pontos para o gr√°fico
    x = np.linspace(-10, 10, 200)
    y = a * x**2 + b * x + c

    # Criar o gr√°fico
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'r-', linewidth=3, label=f'y = {a}x¬≤ + {b}x + {c}')

    # Destacar o v√©rtice
    x_vertice = -b / (2*a)
    y_vertice = a * x_vertice**2 + b * x_vertice + c
    plt.plot(x_vertice, y_vertice, 'go', markersize=10, label=f'V√©rtice: ({x_vertice:.1f}, {y_vertice:.1f})')

    # Configurar o gr√°fico
    plt.title(f'Fun√ß√£o Quadr√°tica: y = {a}x¬≤ + {b}x + {c}', fontsize=14, fontweight='bold')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.axhline(y=0, color='k', linewidth=0.8)
    plt.axvline(x=0, color='k', linewidth=0.8)
    plt.show()

    # Mostrar an√°lise
    print(f"üìä An√°lise da fun√ß√£o y = {a}x¬≤ + {b}x + {c}:")
    print(f"   ‚Ä¢ Coeficiente a: {a} ({'positivo' if a > 0 else 'negativo'} - concavidade {'para cima' if a > 0 else 'para baixo'})")
    print(f"   ‚Ä¢ V√©rtice: ({x_vertice:.2f}, {y_vertice:.2f})")

    if a > 0:
        print(f"   ‚Ä¢ Valor m√≠nimo: y = {y_vertice:.2f}")
    else:
        print(f"   ‚Ä¢ Valor m√°ximo: y = {y_vertice:.2f}")

# Testando nossa fun√ß√£o
grafico_funcao_quadratica(1, -2, -3)
````

### üéØ Exemplo 1: Slider Simples

````python

def funcao_com_slider(a=1):
    '''
    Fun√ß√£o que ser√° controlada por um slider
    '''
    x = np.linspace(-5, 5, 100)
    y = a * x

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, 'b-', linewidth=3, label=f'y = {a}x')
    plt.title(f'Fun√ß√£o Linear: y = {a}x', fontsize=14)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    plt.ylim(-10, 10)
    plt.legend()
    plt.show()

# Criar o elemento interativo
interact(funcao_com_slider,
         a=FloatSlider(min=-3, max=3, step=0.1, value=1, description='Coeficiente a:'))
````


### üéØ Exemplo 2: M√∫ltiplos Controles


````python


def explorar_parabola(a=1, b=0, c=0, mostrar_vertice=True, cor='azul'):
    '''
    Explora par√°bolas com m√∫ltiplos controles
    '''
    x = np.linspace(-10, 10, 200)
    y = a * x**2 + b * x + c

    # Definir cor
    cores = {'azul': 'blue', 'vermelho': 'red', 'verde': 'green', 'roxo': 'purple'}
    cor_grafico = cores[cor]

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, color=cor_grafico, linewidth=3, label=f'y = {a}x¬≤ + {b}x + {c}')

    if mostrar_vertice:
        x_v = -b / (2*a)
        y_v = a * x_v**2 + b * x_v + c
        plt.plot(x_v, y_v, 'ro', markersize=10, label=f'V√©rtice: ({x_v:.1f}, {y_v:.1f})')

    plt.title(f'y = {a}x¬≤ + {b}x + {c}', fontsize=14, fontweight='bold')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.axhline(y=0, color='k', linewidth=0.8)
    plt.axvline(x=0, color='k', linewidth=0.8)
    plt.ylim(-20, 20)
    plt.show()

# Interface interativa completa
interact(explorar_parabola,
         a=FloatSlider(min=-2, max=2, step=0.1, value=1, description='a (x¬≤):'),
         b=FloatSlider(min=-5, max=5, step=0.5, value=0, description='b (x):'),
         c=FloatSlider(min=-10, max=10, step=1, value=0, description='c (constante):'),
         mostrar_vertice=Checkbox(value=True, description='Mostrar v√©rtice'),
         cor=Dropdown(options=['azul', 'vermelho', 'verde', 'roxo'], value='azul', description='Cor:'))
````

### üìö Exemplo: Mini-Aula sobre Fun√ß√£o Exponencial

````python

# ===== INTRODU√á√ÉO =====
print("üéØ MINI-AULA: FUN√á√ÉO EXPONENCIAL")
print("=" * 50)
print("Objetivo: Compreender o comportamento da fun√ß√£o f(x) = a^x")
print("Compet√™ncias BNCC: EM13MAT302, EM13MAT402")
print()

# ===== CONCEITOS TE√ìRICOS =====
print("üìñ CONCEITOS FUNDAMENTAIS")
print("-" * 30)
print("‚Ä¢ Fun√ß√£o exponencial: f(x) = a^x, onde a > 0 e a ‚â† 1")
print("‚Ä¢ Se a > 1: fun√ß√£o crescente")
print("‚Ä¢ Se 0 < a < 1: fun√ß√£o decrescente")
print("‚Ä¢ Dom√≠nio: todos os n√∫meros reais")
print("‚Ä¢ Contradom√≠nio: n√∫meros reais positivos")
print()

# ===== EXEMPLO INTERATIVO =====
def explorar_exponencial(base=2, mostrar_grafico_log=False, incluir_comparacao=True):
    '''
    Explora fun√ß√µes exponenciais interativamente
    '''
    x = np.linspace(-3, 3, 100)
    y = base ** x

    plt.figure(figsize=(12, 6))

    if incluir_comparacao:
        # Mostrar m√∫ltiplas bases para compara√ß√£o
        bases = [0.5, 2, 3]
        cores = ['red', 'blue', 'green']
        for i, b in enumerate(bases):
            y_comp = b ** x
            estilo = '-' if b == base else '--'
            alpha = 1.0 if b == base else 0.5
            plt.plot(x, y_comp, color=cores[i], linestyle=estilo,
                    linewidth=3, alpha=alpha, label=f'y = {b}^x')
    else:
        plt.plot(x, y, 'b-', linewidth=3, label=f'y = {base}^x')

    # Destacar pontos importantes
    plt.plot(0, 1, 'ro', markersize=10, label='(0, 1) - sempre passa aqui!')
    plt.plot(1, base, 'go', markersize=8, label=f'(1, {base})')

    plt.title(f'Fun√ß√£o Exponencial: y = {base}^x', fontsize=14, fontweight='bold')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.axhline(y=0, color='k', linewidth=0.8)
    plt.axvline(x=0, color='k', linewidth=0.8)
    plt.ylim(0, 10)

    if mostrar_grafico_log:
        # Adicionar fun√ß√£o logar√≠tmica (inversa)
        y_log = np.log(x[x > 0]) / np.log(base)
        x_log = x[x > 0]
        plt.plot(x_log, y_log, 'm--', linewidth=2, label=f'y = log_{base}(x)')
        plt.legend()

    plt.show()

    # An√°lise textual
    print(f"üìä AN√ÅLISE DA FUN√á√ÉO y = {base}^x:")
    if base > 1:
        print(f"   ‚úÖ Fun√ß√£o CRESCENTE (base {base} > 1)")
        print("   ‚Ä¢ Quando x aumenta, y aumenta rapidamente")
        print("   ‚Ä¢ Quando x ‚Üí -‚àû, y ‚Üí 0")
        print("   ‚Ä¢ Quando x ‚Üí +‚àû, y ‚Üí +‚àû")
    elif base < 1:
        print(f"   ‚úÖ Fun√ß√£o DECRESCENTE (base {base} < 1)")
        print("   ‚Ä¢ Quando x aumenta, y diminui")
        print("   ‚Ä¢ Quando x ‚Üí -‚àû, y ‚Üí +‚àû")
        print("   ‚Ä¢ Quando x ‚Üí +‚àû, y ‚Üí 0")

    print(f"   ‚Ä¢ Passa sempre pelo ponto (0, 1)")
    print(f"   ‚Ä¢ Passa pelo ponto (1, {base})")
    print(f"   ‚Ä¢ N√£o toca o eixo x (ass√≠ntota horizontal y = 0)")

print("üéÆ EXPLORADOR INTERATIVO:")
interact(explorar_exponencial,
         base=FloatSlider(min=0.1, max=5, step=0.1, value=2, description='Base (a):'),
         mostrar_grafico_log=Checkbox(value=False, description='Mostrar log (fun√ß√£o inversa)'),
         incluir_comparacao=Checkbox(value=True, description='Comparar com outras bases'))
````

````python


def exemplo_multiplos_graficos():
    '''
    Como criar m√∫ltiplos gr√°ficos lado a lado
    '''
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    x = np.linspace(-5, 5, 100)

    # Gr√°fico 1
    ax1.plot(x, x**2, 'b-', linewidth=2)
    ax1.set_title('Fun√ß√£o Quadr√°tica')
    ax1.set_xlabel('x')
    ax1.set_ylabel('y = x¬≤')
    ax1.grid(True, alpha=0.3)

    # Gr√°fico 2
    ax2.plot(x, x**3, 'r-', linewidth=2)
    ax2.set_title('Fun√ß√£o C√∫bica')
    ax2.set_xlabel('x')
    ax2.set_ylabel('y = x¬≥')
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

exemplo_multiplos_graficos()
````

#### 2. **Anota√ß√µes nos Gr√°ficos**


````python


def exemplo_anotacoes():
    '''
    Como adicionar anota√ß√µes e destaques
    '''
    x = np.linspace(-3, 3, 100)
    y = x**2 - 2*x - 3

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'b-', linewidth=3, label='y = x¬≤ - 2x - 3')

    # Encontrar e destacar as ra√≠zes
    raizes_x = [-1, 3]  # Ra√≠zes calculadas
    raizes_y = [0, 0]
    plt.plot(raizes_x, raizes_y, 'ro', markersize=10, label='Ra√≠zes')

    # Destacar o v√©rtice
    x_vertice = 1
    y_vertice = -4
    plt.plot(x_vertice, y_vertice, 'go', markersize=10, label='V√©rtice')

    # Adicionar anota√ß√µes
    plt.annotate('Raiz: (-1, 0)', xy=(-1, 0), xytext=(-2, 2),
                arrowprops=dict(arrowstyle='->', color='red'),
                fontsize=12, color='red')

    plt.annotate('V√©rtice: (1, -4)', xy=(1, -4), xytext=(2, -2),
                arrowprops=dict(arrowstyle='->', color='green'),
                fontsize=12, color='green')

    plt.title('Par√°bola com Pontos Importantes Destacados', fontsize=14, fontweight='bold')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.axhline(y=0, color='k', linewidth=0.8)
    plt.axvline(x=0, color='k', linewidth=0.8)
    plt.show()

exemplo_anotacoes()
````

## üìö Parte 7: Exemplos Pr√°ticos para Diferentes N√≠veis

### üéØ Exemplo para 9¬∫ Ano: Fun√ß√£o Linear Simplificada

Vamos criar um material adequado para alunos que est√£o vendo fun√ß√£o linear pela primeira vez:


````python
def funcao_linear_9ano(coeficiente_angular=1, termo_independente=0,
                       mostrar_pontos=True, mostrar_tabela=False):
    '''
    Fun√ß√£o linear adaptada para 9¬∫ ano - linguagem mais simples
    '''
    # Criar pontos para o gr√°fico
    x = np.linspace(-5, 5, 100)
    y = coeficiente_angular * x + termo_independente

    plt.figure(figsize=(10, 7))
    plt.plot(x, y, 'b-', linewidth=4, label=f'y = {coeficiente_angular}x + {termo_independente}')

    if mostrar_pontos:
        # Mostrar alguns pontos espec√≠ficos
        x_pontos = np.array([-3, -1, 0, 1, 3])
        y_pontos = coeficiente_angular * x_pontos + termo_independente
        plt.plot(x_pontos, y_pontos, 'ro', markersize=8)

        # Anotar pontos importantes
        plt.plot(0, termo_independente, 'go', markersize=12,
                label=f'Corta o eixo y em: (0, {termo_independente})')

    plt.title(f'Fun√ß√£o do 1¬∫ grau: y = {coeficiente_angular}x + {termo_independente}',
              fontsize=16, fontweight='bold')
    plt.xlabel('x (entrada)', fontsize=14)
    plt.ylabel('y (sa√≠da)', fontsize=14)
    plt.grid(True, alpha=0.4, linewidth=1)
    plt.legend(fontsize=12)
    plt.axhline(y=0, color='k', linewidth=1)
    plt.axvline(x=0, color='k', linewidth=1)
    plt.ylim(-10, 10)
    plt.show()

    # Explica√ß√£o simples
    print("üîç O QUE OBSERVAMOS:")
    print("=" * 40)

    if coeficiente_angular > 0:
        print("üìà A reta est√° SUBINDO (crescente)")
        print(f"   ‚Üí Para cada 1 que x aumenta, y aumenta {coeficiente_angular}")
    elif coeficiente_angular < 0:
        print("üìâ A reta est√° DESCENDO (decrescente)")
        print(f"   ‚Üí Para cada 1 que x aumenta, y diminui {abs(coeficiente_angular)}")
    else:
        print("‚û°Ô∏è A reta est√° na HORIZONTAL (constante)")
        print("   ‚Üí y sempre vale a mesma coisa, n√£o importa o valor de x")

    print(f"üìç A reta corta o eixo y no ponto (0, {termo_independente})")

    if mostrar_tabela:
        print("\nüìä TABELA DE VALORES:")
        print("-" * 20)
        print("x  |  y")
        print("-" * 20)
        for x_val in [-2, -1, 0, 1, 2]:
            y_val = coeficiente_angular * x_val + termo_independente
            print(f"{x_val:2} | {y_val:3}")

print("üéÆ EXPLORE A FUN√á√ÉO DO 1¬∫ GRAU:")
print("(Adequado para 9¬∫ ano)")

interact(funcao_linear_9ano,
         coeficiente_angular=FloatSlider(min=-3, max=3, step=0.5, value=1,
                                       description='Inclina√ß√£o da reta:'),
         termo_independente=FloatSlider(min=-5, max=5, step=1, value=0,
                                      description='Onde corta o eixo y:'),
         mostrar_pontos=Checkbox(value=True, description='Mostrar pontos importantes'),
         mostrar_tabela=Checkbox(value=False, description='Mostrar tabela de valores'))
````
